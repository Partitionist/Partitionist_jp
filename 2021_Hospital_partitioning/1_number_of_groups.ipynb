# 病院数（グループ数）を決定するための参照データを作成

- データクラスタリングのアルゴリズムとして、K-meansとWard法を使用

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'IPAGothic'
cmap = plt.get_cmap("tab10")

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering

prefecture = 'ibaraki'
df = pd.read_excel('dataset/'+prefecture+'/hospitals_'+prefecture+'.xlsx', index_col=0)
X = df[['latitude','longitude']].to_numpy()

n_clusters_list = np.arange(10,100,4, dtype=int)
```

```python
df_doctors_kmeans = pd.DataFrame( columns=['n_clusters','group','n_doctors','district_std'] )
df_doctors_ward = pd.DataFrame( columns=['n_clusters','group','n_doctors','district_std'] )

for n_clusters in n_clusters_list:

    # k-means
    df_ = df.copy()
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
    labels = kmeans.labels_
    df_['kmeans_label'] = [str(label) for label in labels]

    n_doctors_list = df_.groupby(['kmeans_label']).sum()['num_doctors'].tolist()
    df_std = df_.groupby(['kmeans_label']).std()
    district_std = [max(row['latitude'], row['longitude']) for idx,row in df_std.iterrows()]
    group_idx = 0
    for n_doctors in n_doctors_list:
        tmp_se = pd.Series( [ str(n_clusters), group_idx, n_doctors, district_std[group_idx] ], index=df_doctors_kmeans.columns )
        df_doctors_kmeans = df_doctors_kmeans.append( tmp_se, ignore_index=True )
        group_idx += 1
        
    # Ward algorithm
    df_ = df.copy()
    clustering = AgglomerativeClustering(n_clusters=n_clusters).fit(X)
    labels = clustering.labels_
    df_['Ward_label'] = [str(label) for label in labels]

    n_doctors_list = df_.groupby(['Ward_label']).sum()['num_doctors'].tolist()
    df_std = df_.groupby(['Ward_label']).std()
    district_std = [max(row['latitude'], row['longitude']) for idx,row in df_std.iterrows()]
    group_idx = 0
    for n_doctors in n_doctors_list:
        tmp_se = pd.Series( [ str(n_clusters), group_idx, n_doctors, district_std[group_idx] ], index=df_doctors_ward.columns )
        df_doctors_ward = df_doctors_ward.append( tmp_se, ignore_index=True )
        group_idx += 1
        
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,4))

sns.stripplot(x="n_clusters", y="n_doctors", data=df_doctors_kmeans, jitter=0.01, color=cmap(0), ax=ax1)
sns.stripplot(x="n_clusters", y="n_doctors", data=df_doctors_ward, jitter=0.01, color=cmap(1), ax=ax2)

ax1.set_title('K-means')
ax2.set_title('Ward algorithm')
plt.gca().set_yscale('log')

ax1.set_xlabel('病院数')
ax1.set_ylabel('各病院の医師人数')
ax2.set_xlabel('病院数')
ax2.set_ylabel('各病院の医師人数')
plt.tight_layout()
plt.savefig('results/'+prefecture+'/'+prefecture+'_num_doctors.pdf')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,4))

sns.stripplot(x="n_clusters", y="district_std", data=df_doctors_kmeans, jitter=0.01, color=cmap(0), ax=ax1)
sns.stripplot(x="n_clusters", y="district_std", data=df_doctors_ward, jitter=0.01, color=cmap(1), ax=ax2)

ax1.set_title('K-means')
ax2.set_title('Ward algorithm')
plt.gca().set_yscale('log')

ax1.set_xlabel('病院数')
ax1.set_ylabel('統合される病院の広がり（標準偏差）')
ax2.set_xlabel('病院数')
ax2.set_ylabel('統合される病院の広がり（標準偏差）')
plt.tight_layout()
plt.savefig('results/'+prefecture+'/'+prefecture+'_merged_areas.pdf')
```
